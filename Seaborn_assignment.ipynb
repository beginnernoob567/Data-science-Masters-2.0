# Que 1:
    Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
iris=sns.load_dataset('iris')
# 1) Scatter plot
sns.scatterplot(x=iris.sepal_length,y=iris.sepal_width)
```

```python
# 2) Distribution plot
sns.displot(iris['sepal_length'])
```

```python
# 3) Line plot
sns.lineplot(x=iris.sepal_length,y=iris.sepal_width)
```

```python
# 4) Box plot
sns.boxplot(iris.sepal_length)
```

```python
# 5) Barplot
sns.barplot(x=iris.sepal_length,y=iris.sepal_width)
```

# Que 6: 
    Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights=sns.load_dataset('flights')
def monthTOnum(x):
    if x=='Jan':
        return int('01')
    elif x=='Feb':
        return int('02')
    elif x=='Mar':
        return int('03')
    elif x=='Apr':
        return int('04')
    elif x=='May':
        return int('05')
    elif x=='Jun':
        return int('06')
    elif x=='Jul':
        return int('07')
    elif x=='Aug':
        return int('08')
    elif x=='Sep':
        return int('09')
    elif x=='Oct':
        return int('10')
    elif x=='Nov':
        return int('11')
    elif x=='Dec':
        return int('12')
flights['month']=flights['month'].apply(monthTOnum)
```

```python
sns.heatmap(flights)
```
