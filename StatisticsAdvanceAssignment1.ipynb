# Q3.
Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.

```python
import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution.

    Parameters:
    - x: The point at which to evaluate the PDF.
    - mean: The mean of the normal distribution.
    - std_dev: The standard deviation of the normal distribution.

    Returns:
    - PDF value at the given point.
    """
    coefficient = 1 / math.sqrt(2 * math.pi * std_dev**2)
    exponent = -((x - mean)**2) / (2 * std_dev**2)
    pdf_value = coefficient * math.exp(exponent)
    return pdf_value

# Example usage:
mean_value = 0
std_deviation = 1
point = 1
result = normal_pdf(point, mean_value, std_deviation)
print(f"The Normal PDF at point {point} with mean {mean_value} and standard deviation {std_deviation} is: {result}")
```

# Q6.
Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.

```python

import matplotlib.pyplot as plt
import numpy as np
from math import exp, factorial

def poisson_cdf(k, lmbda):
    cdf = 0.0
    for i in range(k + 1):
        cdf += (exp(-lmbda) * lmbda**i) / factorial(i)
    return cdf

# Parameters
mean_value = 3.0

# Generate x values (points to evaluate CDF)
x_values = np.arange(0, 10)

# Calculate CDF values for each x
cdf_values = [poisson_cdf(k, mean_value) for k in x_values]

# Plot the CDF
plt.plot(x_values, cdf_values, marker='o', linestyle='-', color='b')
plt.title(f'Poisson CDF with mean {mean_value}')
plt.xlabel('Number of Events (k)')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()

```

# Q7. 
How Binomial distribution different from Poisson distribution?

### Poisson distribution
        -The poisson distribution describes the number of events that occur in a fixed interval of time.
    
        - Poisson distribution is used when the events are random and no fixed number of trials are required.
         
### Binomial distribution
        -Binomial ditribution describes the number of successes that occur from a fixed number of bernoulli trials.
        
        - Binomial distribution is used when number of events are fixed and the number of trials is also fixed
        

# Q8. 
Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
sample mean and variance.

```python
import numpy as np

# Set the seed for reproducibility
np.random.seed(42)

# Parameters for the Poisson distribution
mean_value = 5
sample_size = 1000

# Generate a random sample from a Poisson distribution
poisson_sample = np.random.poisson(mean_value, size=sample_size)

# Calculate the sample mean and variance
sample_mean = np.mean(poisson_sample)
sample_variance = np.var(poisson_sample)

print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```

# Q10. 
In normal distribution with respect to mean position, where does the least frequent data appear?

```python
# In normal ditribution the least frequent data appears on the tails of the distribution.
# For example in the following normal distribution
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(42)

# Parameters for the normal distribution
mean_value = 0
std_deviation = 1
sample_size = 1000

# Generate a random sample from a normal distribution
normal_sample = np.random.normal(mean_value, std_deviation, size=sample_size)

# Plot a histogram of the generated sample
plt.hist(normal_sample, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')

# Plot the probability density function (PDF) for comparison
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pdf = (1/(std_deviation * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean_value)/std_deviation)**2)
plt.plot(x, pdf, 'k', linewidth=2)

# Add labels and title
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')

# Show the plot
plt.show()
```
