Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

```python
# Matplotlib is a python library that is used to plot graphs and visualise data
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# types of plots include:
# i) Scatter plot
print("Scatter plot: \n",plt.scatter(x,y))
plt.show()
# ii) line plot
print("Line plot: \n",plt.plot(x,y))
plt.show()
# iii) Bar plot
print("bar plot: \n",plt.bar(y,x))
plt.show()
# iv) Box plot
print("Box plot: \n",plt.boxplot(y,vert=0))
plt.show()
# v) Histogram
print("Histogram: \n",plt.hist(x))
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# A bar plot or bar chart is a graph that represents the category of data with rectangular
# bars with lengths and heights that is proportional to the values which they represent. 
plt.subplot(2,1,1)
plt.bar(company,profit)
# Horizontal bar graph
plt.subplot(2,1,2)
plt.barh(company,profit)
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
# In the box plot, a box is created from the first quartile to the third quartile, 
# a vertical line is also there which goes through the box at the median. 
# Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.

# A Box plot is created to display the summary of the set of data values having 
# properties like minimum, first quartile, median, third quartile and maximum.
print("Plotting box_plot1: ")
plt.boxplot(box1)
plt.show()
print("Plotting box_plot2: ")
plt.boxplot(box2)
plt.show()
```
