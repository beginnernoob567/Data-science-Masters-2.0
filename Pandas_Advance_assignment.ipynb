Q1. List any five functions of the pandas library with execution.

```python
import pandas as pd
# 1) DataFrame function creates a DataFrame
dict1={"Key1":[1,2,3,4,5],
       "Key2":[6,7,8,9,0],
       "Key3":[23,564,7,7,9]}
df1=pd.DataFrame(dict1)
print(df1)
print(type(df1))
```

```python
# 2) Series function creates a series
series=[1,2,3,44,56,76,78]
s1=pd.Series(series)
print(s1)
print(type(s1))
```

```python
# 3) to_datetime converts the object to datetime datatype
dt=['2022-04-23','2023-05-15','2022-06-13','1998-04-16','1999-11-23']
dict1={"Key1":[1,2,3,4,5],
       "Key2":[6,7,8,9,0],
       "Key3":[23,564,7,7,9],
       "Date":dt}
df1=pd.DataFrame(dict1)
df1['New_Date']=pd.to_datetime(df1['Date'])
print(df1.dtypes)
```

```python
# 4) Rolling goes from one row to another
#    Window specifies number of elements to consider
dict3={"A":[1,2,3,4,5,7,8,9,1,2],
       "B":[6,7,8,9,0,4,5,6,1,2],
       "Values":[23,564,7,7,9,3,4,67,89,78]}

df3=pd.DataFrame(dict3)
print(df3['Values'].rolling(window=3).sum())
```

```python
# 5) reindex resets the index of the dataframe
dict2={"A":[1,2,3,4,5],
       "B":[6,7,8,9,0],
       "C":[23,564,7,7,9]}
df2=pd.DataFrame(dict2)
print(df2)
df2=df2.reindex([i for i in range(1,len(df2)*2,2)])
print(df2)
```

Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column
'Word_Count' that contains the number of words in each row of the 'Text' column.

```python
dict4={"A":[1,2,3,4,5],
       "B":[6,7,8,9,0],
       "Values":[23,564,7,7,5],
       "Text":['I hope this turns into something great.','It’s late at night and everyone is asleep.','My mind like my body is twisting, turning, restless.','I guess I never really thought about it.','The first thing I remembered this morning after I woke up was my purple nose in my dream.']}
df4=pd.DataFrame(dict4)
print(df4)
def word_count(word):
    count=0
    for i in word.split():
        count+=1
    return count
df4['Word_count']=df4['Text'].apply(word_count)
print("\n",df4)
```

Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email
addresses in the format 'username@domain.com'. Write a Python function that creates a new column
'Username' in df that contains only the username part of each email address.


The username is the part of the email address that appears before the '@' symbol. For example, if the
email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your
function should extract the username from each email address and store it in the new 'Username'
column.

```python
dict5={"A":[1,2,3,4],
       "B":[6,7,8,9],
       "Values":[23,564,7,7],
       "Email":['john.doe@domain.com','jack.daniels@domain.com','horacio.pagani@domain.com','ratan.tata@domain.com']}
df5=pd.DataFrame(dict5)
print(df5)
def extract(x):
    user=[]
    separator=''
    for i in x:
        if i=='@':
            break
        elif i=='.':
            
            user.append(' ')
        else:
           user.append(str(i))
    return separator.join(user)
df5['Username']=df5['Email'].apply(extract)
print("\n",df5,'\n')
```

Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new
column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g.
Monday, Tuesday) corresponding to each date in the 'Date' column.

For example, if df contains the following values:

Date

0 2023-01-01

1 2023-01-02

2 2023-01-03

3 2023-01-04

4 2023-01-05


Your function should create the following DataFrame:

Date Weekday

0 2023-01-01 Sunday

1 2023-01-02 Monday

2 2023-01-03 Tuesday

3 2023-01-04 Wednesday

4 2023-01-05 Thursday

The function should return the modified DataFrame.

```python
df10=pd.DataFrame({'Sales':['2000','3000','1500','6000','5000','7000','10000','1000','3300','4500','2600','8000','5500','11000'],
        'Date':['2022-03-22', '2022-03-23', '2022-03-24', '2022-03-25',
               '2022-03-26', '2022-03-27', '2022-03-28', '2022-03-29',
               '2022-03-30', '2022-03-31', '2022-04-01', '2022-04-02',
               '2022-04-03', '2022-04-04']})

df10['Date']=pd.to_datetime(df10['Date'])
def weekday(x):
    if x==0:
        return 'Monday'
    elif x==1:
        return 'Tuesday'
    elif x==2:
        return 'Wednesday'
    elif x==3:
        return 'Thursday'
    elif x==4:
        return 'Friday'
    elif x==5:
        return 'Saturday'
    else:
        return 'Sunday'
df11=pd.DataFrame({'Date':df10['Date'],
                   'Weekday':df10['Date'].dt.weekday.apply(weekday)})
print(df11)
```
